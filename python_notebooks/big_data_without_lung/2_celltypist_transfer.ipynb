```python
import scanpy as sc
import celltypist
import time
import numpy as np
import mygene
```

```python
fallopian_tube = sc.read('celltypist_demo_folder/Fallopian tube_normal_1.h5ad', backup_url = 'https://datasets.cellxgene.cziscience.com/116cae82-6420-48a8-af2f-bc9f64545788.h5ad')
```

```python
kidney_tube = sc.read('celltypist_demo_folder/Kidney_normal_fetal.h5ad', backup_url = 'https://datasets.cellxgene.cziscience.com/bae8bc5b-de5a-470b-9fd6-5dc4e0e750cf.h5ad')
```

```python
kidney_tube.obs['cell_type'].value_counts()
```

```python
kidney_tube.var
```

```python
# Get current indices from adata.var (list of ENSG IDs)
ensg_ids = kidney_tube.var.index.tolist()

# Initialize mygene object
mg = mygene.MyGeneInfo()

# Query gene symbols for all ENSG IDs
gene_info = mg.querymany(ensg_ids, scopes='ensembl.gene', fields='symbol', species='human')

# Create mapping dictionary for ENSG -> symbol conversion
gene_mapping = {item['query']: item.get('symbol', None) for item in gene_info}
```

```python
# Create new index based on gene_mapping
new_index = [gene_mapping.get(ensg, ensg) for ensg in kidney_tube.var.index]

# Update the index in adata.var with gene symbols
kidney_tube.var.index = new_index
```

```python
kidney_tube = kidney_tube[:, kidney_tube.var.index.notnull()]
```

```python
sc.pp.normalize_total(kidney_tube, target_sum = 1e4)
sc.pp.log1p(kidney_tube)
```

```python
# 11 cell types in the first data
kidney_tube.obs.cell_type.unique()
```

```python
kidney_tube.shape
```

```python
# Sample 500 cells from each cell type for `adata_Elmentaite`
# All cells from a given cell type will be selected if the cell type size is < 500
sampled_cell_index = celltypist.samples.downsample_adata(kidney_tube, mode = 'each', n_cells = 1000, by = 'cell_type', return_index = True)
```

```python
print(f"Number of downsampled cells for training: {len(sampled_cell_index)}")
```

```python
# You can also set `mini_batch = True` to enable mini-batch training.
t_start = time.time()
model_fs = celltypist.train(kidney_tube[sampled_cell_index], 'cell_type', n_jobs = 10, max_iter = 5, use_SGD = True)
t_end = time.time()
print(f"Time elapsed: {t_end - t_start} seconds")
```

```python
gene_index = np.argpartition(np.abs(model_fs.classifier.coef_), -100, axis = 1)[:, -100:]
```

```python
gene_index = np.unique(gene_index)
```

```python
print(f"Number of genes selected: {len(gene_index)}")
```

```python
# Add `check_expression = False` to bypass expression check with only a subset of genes.
t_start = time.time()
model = celltypist.train(kidney_tube[sampled_cell_index, gene_index], 'cell_type', check_expression = False, n_jobs = 10, max_iter = 100)
t_end = time.time()
print(f"Time elapsed: {(t_end - t_start)/60} minutes")
```

```python
model.write('celltypist_demo_folder/Kidney_normal_fetal.pkl')
```
